> ##### load to biz_log_summary data from postgres

```python
import psycopg2
import pickle

db_info = {
    'host'     : "192.168.0.175",
    'port'     : 5432,
    'database' : "psm",
    'user'     : "psm",
    'password' : "psm123"
}

conn = psycopg2.connect( **db_info )

cur = conn.cursor()

sql = \
'''
SELECT log_seq, proc_date, proc_time, emp_user_id, emp_user_name, dept_id, dept_name, user_ip, user_id, scrn_id, scrn_name, req_type, system_seq, server_seq, req_context, req_url, req_end_time, result_type, result_type_ko, query_id, rd_id, rd_name, grid_id, grid_name, request_data
FROM public.biz_log_summary;
'''

print("The number of parts: ", cur.rowcount)

cur.execute( sql )

rows = []
row = cur.fetchone()
rows.append( row )
# print( type( row ) )

while row is not None:
    row = cur.fetchone()
    rows.append( row )

with open('./biz_log_summary.pickle','wb') as f:
    pickle.dump( rows, f )


print( len(rows) )
# display the PostgreSQL database server version
# db_version = cur.fetchone()
# print(db_version)

# close the communication with the PostgreSQL
cur.close()
```
