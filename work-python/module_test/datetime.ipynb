> ##### auto generate time series data Funciton

```python
import pandas as pd

def generate_time_series( 
                         start_d : str, end_d : str, 
                         start_t : int, end_t : int, 
                         period : str
                         ):
    series = pd.date_range( start=start_d, end=end_d, freq='D' )
    
    df = pd.DataFrame( columns=['ds','values'])
    df.ds, df.values = series, 0
    
    tick = df.set_index('ds').resample( period ).count().reset_index()
    
    tick.values = np.random.randint( 10, 20, size=(len(tick.ds)))
    
    tick['days'] = tick['ds'].dt.day_name()
    tick['hour'] = tick['ds'].dt.hour
    
    tick.loc[ (tick.days == 'Saturday') | (tick.days == 'Sunday'),  'values' ] = 0
    tick.loc[ (tick.hour < 9 ) | (tick.hour > 18 ), 'values' ] = 0
    
    return tick

gen_df = generate_time_series(
        start_d= '2020-01-01', end_d= '2020-12-31', 
        start_t= 9, end_t= 18, 
        period= '1T'
)

gen_df    
```
