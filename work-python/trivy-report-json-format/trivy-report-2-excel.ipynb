```python
import pandas as pd
import json
import os

xlsx_filename = f'trivy-report.xlsx'
base_path = f'./data/'

error_file = []

for filename in os.listdir( base_path ):
    sheetname = filename.replace(".json","")

    xlsx_filename = f'trivy-report-{filename}.xlsx'

    if( filename.endswith( 'json') ):
        with open( f'{base_path}{filename}', 'r' ) as fr:
            
            data = json.loads( fr.read() )   

        try:
            print( f'filename : {filename}')
            os_flag = False
            jar_flag = False
            
            if 'Vulnerabilities' in data.get('Results')[0].keys():
                os_image = data.get('Results')[0].get('Vulnerabilities')
                os_flag = True
            else:
                os_flag = False

            if 'Vulnerabilities' in data.get('Results')[1].keys():
                jar = data.get('Results')[1].get('Vulnerabilities')
                jar_flag = True
            else:
                jar_flag = False

            os_df = ''
            jar_df = ''
            os_df = pd.DataFrame( os_image )
            jar_df = pd.DataFrame( jar )

            if os.path.exists( xlsx_filename ):
                with pd.ExcelWriter( f'./data/{xlsx_filename}', mode='A' ) as writer:
                    if( os_flag ):
                        os_df.to_excel(writer, sheet_name = f'{sheetname}-os')
                    if( jar_flag ):
                        jar_df.to_excel(writer, sheet_name = f'{sheetname}-jar') 
            else:
                with pd.ExcelWriter( f'./data/{xlsx_filename}', mode='w' ) as writer:
                    if( os_flag ):
                        os_df.to_excel(writer, sheet_name = f'{sheetname}-os')
                    if( jar_flag ):
                        jar_df.to_excel(writer, sheet_name = f'{sheetname}-jar') 

            print( data.get('Results')[0].get('Vulnerabilities') )
        except:
            print( f'Error : {xlsx_filename}')
            error_file.append({
                'name' : f'filename',
                'os': os_df,
                'jar_df': jar_df
            })
```
